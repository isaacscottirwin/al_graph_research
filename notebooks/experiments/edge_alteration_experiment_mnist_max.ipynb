```python
from IPython.display import HTML
from al_graph_research.data.datasets import MnistDataset
import al_graph_research.experiments.only_edges.visualizations as viz_only_edges
import al_graph_research.experiments.only_edges.configurations as configs
```

```python

dataset = MnistDataset(4, 9, metric="vae", num_samples_per_class=300, seed=42)
```

```python
config_zero = configs.ExperimentConfigurations.MNIST_ZERO_MAX
config_negate = configs.ExperimentConfigurations.MNIST_NEGATE_MAX
exp_zero = configs.ExperimentConfigBuilder.build_experiment(config_zero, seed=42)
exp_negate = configs.ExperimentConfigBuilder.build_experiment(config_negate, seed=42)
```

```python
result_zero = exp_zero.run(dataset)
result_negate = exp_negate.run(dataset)
```

```python
viz_only_edges.plot_metric_comparison(
    [result_zero, result_negate],
    "accuracy_history",
    ["Delete edges", "Negate edges"],
    title="Accuracy comparison"
)

viz_only_edges.plot_metric_comparison(
    [result_zero, result_negate],
    "lam2_history",
    ["Delete edges", "Negate edges"],
    title="λ₂ comparison"
)

```

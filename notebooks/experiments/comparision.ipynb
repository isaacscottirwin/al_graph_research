```python
import numpy as np
from IPython.display import HTML
from al_graph_research.experiments.only_edges.edge_alteration_experiment import EdgeAlterationExperiment
from al_graph_research.data.datasets import GaussianDataset, MnistDataset
import al_graph_research.experiments.only_edges.visualizations as viz_only_edges
import al_graph_research.visualization.visualizations as viz
from al_graph_research.graphs.knn_graph import KNNGraph
from al_graph_research.graphs.signed_laplacian import SignedLaplacian
```

```python
mean1 = [.5, .4]
mean2 = [.9, -1.7]
mean3 = [-.4, .5]
mean_arr = [mean1, mean2, mean3]
cov = [[.1, 0], [0, .1]]
dataset_one = GaussianDataset(
    n_per_cluster=500,
    cov=cov,
    mean_arr=mean_arr,
    n_blobs=3,
    seed=42,
)

dataset_two = MnistDataset(0, 1, metric="vae", num_samples_per_class=500, seed=42)

```

```python
graph_legacy = KNNGraph(20, "uniform", dataset_one.data, construction_method="legacy")
graph_partner = KNNGraph(20, "uniform", dataset_one.data, construction_method="partner")

A_legacy = graph_legacy.A_sp
A_partner = graph_partner.A_sp

print("legacy nnz:", A_legacy.nnz)
print("partner nnz:", A_partner.nnz)
print("different entries:", (A_legacy != A_partner).nnz) # type: ignore
print("max abs diff:", np.max(np.abs((A_legacy - A_partner).data)) if (A_legacy - A_partner).nnz > 0 else 0.0)
```

```python
graph_legacy = KNNGraph(20, "uniform", dataset_two.data, construction_method="legacy")
graph_partner = KNNGraph(20, "uniform", dataset_two.data, construction_method="partner")

A_legacy = graph_legacy.A_sp
A_partner = graph_partner.A_sp

print("legacy nnz:", A_legacy.nnz)
print("partner nnz:", A_partner.nnz)
print("different entries:", (A_legacy != A_partner).nnz) # type: ignore
print("max abs diff:", np.max(np.abs((A_legacy - A_partner).data)) if (A_legacy - A_partner).nnz > 0 else 0.0)
```
